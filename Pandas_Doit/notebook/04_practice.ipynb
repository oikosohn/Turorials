# 한 번에 4개의 그래프 그리기

## matplotlib 라이브러리로 그래프 그리기
1. 전체 그래프가 위치할 기본 틀 만들기
2. 그래프를 그려 넣을 그래프 격자 만들기
3. 격자에 그래프를 하나씩 추가하기, 격자에 그래프가 추가되는 순서는 왼쪽에서 오른쪽 방향입니다.
4. 만약 격자의 첫 번재 행이 꽉 차면 두 번재 행에 그래프를 그려 넣습니다.

```python
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
```

```python
# 그래프 격자가 위치할 기본 틀 만들기
fig = plt.figure()

# 그래프 격자 그리기
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
```

```python
# 격자에 그래프 전달
axes1.plot(dataset_1['x'],dataset_1['y'],'o')
axes2.plot(dataset_2['x'],dataset_2['y'],'o')
axes3.plot(dataset_3['x'],dataset_3['y'],'o')
axes4.plot(dataset_4['x'],dataset_4['y'],'o')
fig
```

```python
axes1.set_title('dataset_1')
axes2.set_title('dataset_2')
axes3.set_title('dataset_3')
axes4.set_title('dataset_4')
fig
```

```python
fig.suptitle('Anscombe Data')
fig
```

```python
fig.tight_layout()
fig
```

데이터셋 1,2,3,4의 평균, 분산, 상관관계, 회귀선의 통계 수치가 같으나 그래프의 형태는 다르다.

# 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

```python
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

## 히스토그램

히스토그램 : 데이터프레임의 열 데이터 분포와 빈도를 살펴보는 용도

변수 : total_bill, tip 등의 열을 변수라고 부르기도 한다.

일변량 그래프(단변량 그래프) : 변수를 하나만 사용해서 그린 그래프를 뜻하며 히스토그램 역시 그러하다.

```python
fig = plt.figure()
axe1=fig.add_subplot(1,1,1)
```

```python
axe1.hist(tips['total_bill'],bins=10)
axe1.set_title('Histogram of Total Bill')
axe1.set_xlabel('Frequency')
axe1.set_ylabel('Total Bill')
fig
```

## 산점도

산점도 그래프 : 변수를 2개 사용해서 통계 용어로 이변량 그래프라고 부른다.

```python
scatter_plot = plt.figure()
axe1 = scatter_plot.add_subplot(1,1,1)
axe1.scatter(tips['total_bill'], tips['tip'])
axe1.set_title('Scatterplot of Total Bill vs Tip')
axe1.set_xlabel('Total Bill')
axe1.set_ylabel('Tip')
```

## 박스 그래프

박스 그래프 : 이산형 변수와 연속형 변수를 함께 사용하는 그래프

```python
boxplot = plt.figure()
axe1 = boxplot.add_subplot(1,1,1)
axe1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
               labels=['Female','Male']
              )
axe1.set_xlabel('Sex')
axe1.set_ylabel('Tip')
axe1.set_title('Boxplot of Tips by Sex')
```

## 다변량 그래프

다변량 그래프 : 3개 이상의 변수를 사용한 그래프

### 산점도 그래프

```python
def recode_sex(sex):
    if sex=='Female':
        return 0
    else:
        return 1
```

apply메서드 : 함수를 브로드캐스팅할 때 사용한다.

```python
tips['sex_color']=tips['sex'].apply(recode_sex)
```

```python
scatter_plot=plt.figure()
axe1=scatter_plot.add_subplot(1,1,1)
axe1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,
    c=tips['sex_color'],
    # 점의 투명도 설정
    alpha=0.5)
axe1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axe1.set_xlabel('Total Bill')
axe1.set_ylabel('Tip')
```

점의 크기 : 테이블당 인원
점의 색상 : 남녀 구분

# 단변량 그래프 그리기 - 히스토그램

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

```python
ax = plt.subplots()

# distplot 메서드에 열 데이터 전달
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax = plt.subplots()

# ked=False : 밀집도 그래프 제외
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Ttoal Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

아래의 밀집도 그래프는 주어진 데이터를 정규화시켜 넓이가 1이 되도록 그린 그래프다.

```python
ax = plt.subplots()

# hist=False : 히스토그램 제외
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Ttoal Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

양탄자 그래프 : 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집 정도를 표현한 그래피이다.

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

## count 그래프
count 그래프 : 이산값을 표현한 그래프이다.

```python
ax = plt.subplots()

# tips 데이터프레임의 day 열 데이터를 입력
ax = sns.countplot('day',data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

## seaborn 라이브러리로 산점도 그래프 그리기

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill',y='tip',data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill',y='tip',data=tips, fit_reg=False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

jointplot 메서드 : 그래프 이름과 x,y 축 이름을 추가할 수 있다.

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

육각 그래프

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
joint.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

이차원 밀집도

```python
ax=plt.subplots()
ax=sns.kdeplot(data=tips['total_bill'],
              data2=tips['tip'],
              shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

pairplot의 중복 문제 해결

```python
pair_grid=sns.PairGrid(tips)

# 윗삼각형 : 회귀선을 포함한 산점도
pair_grid=pair_grid.map_upper(sns.regplot)

# 아랫삼각형 : 이차원 밀집도
pair_grid=pair_grid.map_lower(sns.kdeplot)

# 대각선 : 히스토그램, 밀집도
pair_grid=pair_grid.map_diag(sns.distplot, rug=True)

plt.show()
```

# seaborn 라이브러리로 그래프 스타일 설정하기

```python
sns.set_style('whitegrid')
fig,ax=plt.subplots()
ax=sns.violinplot(x='time',y='total_bill',hue='sex',data=tips, split=True)
```

```python
fig=plt.figure()
seabron_style = ['darkgrid','whitegrid','dark','white','ticks']

for idx, style in enumerate(seabron_style):
    plot_position = idx+1
    with sns.axes_style(style):
        # 2행 2열
        ax = fig.add_subplot(2,3,plot_position)
        
        violin = sns.violinplot(x='time',y='total_bill',data=tips, ax=ax)
        violin.set_title(style)
fig.tight_layout()
```
