## Build hypothesis and cost 

```
import tensorflow as tf
```

```
x_data = [1,2,3,4,5]
y_data = [1,2,3,4,5]
w = tf.Variable(2.9) # w 초기값 : 보통 초기에 랜덤하게 부여한다.
b = tf.Variable(0.5) # b 초기값 : 보통 초기에 랜덤하게 부여한다.

# hypothesis = w*x + b
hypothesis = w * x_data + b
cost = tf.reduce_mean(tf.square(hypothesis-x_data))
```

```
# reduce_mean : 차원(랭크) 줄이면서 평균 구한다.
v = [1,2,3,4]
tf.reduce_mean(v)
```

```
tf.square(3)
```

```
learning_rate = 0.01
```

```
for i in range(100+1):
  # Gradient decent
  with tf.GradientTape() as tape:
    hypothesis = w * x_data + b
    cost = tf.reduce_mean(tf.square(hypothesis-y_data))
  w_grad, b_grad = tape.gradient(cost, [w,b])
  w.assign_sub(learning_rate*w_grad)
  b.assign_sub(learning_rate*b_grad)
  if i%10 == 0:
    print('{:5}번{:10.4f}{:10.4f}{:10.6f}'.format(i, w.numpy(), b.numpy(), cost))
```
